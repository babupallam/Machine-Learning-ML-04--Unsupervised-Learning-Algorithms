#  K-Means Clustering

### Overview of K-Means Clustering

**K-Means Clustering** is an unsupervised machine learning algorithm primarily used for partitioning a dataset into a set of distinct, non-overlapping groups or clusters. It aims to partition the data points into \( k \) clusters, where each data point belongs to the cluster with the nearest mean, serving as a prototype of the cluster.

### Key Concepts

1. **Clusters**: A cluster is a collection of data points aggregated together because of certain similarities.
  
2. **Centroid**: The centroid of a cluster is the mean position of all the points in the cluster. It serves as the representative of that cluster.

3. **Distance Metric**: Typically, the Euclidean distance is used to measure the similarity (or dissimilarity) between data points. The distance determines how close a point is to the centroid of a cluster.

4. **Number of Clusters (k)**: The algorithm requires the number of clusters \( k \) to be predefined. The value of \( k \) significantly impacts the clustering result and is often chosen based on domain knowledge, heuristics, or through methods like the Elbow method.

### Algorithmic Steps

The K-Means algorithm follows these steps:

1. **Initialization**: Choose \( k \) initial centroids. These can be chosen randomly, or by some heuristic like the K-Means++ method, which ensures better starting points.

2. **Assignment Step**:
   - For each data point, calculate the distance to each of the \( k \) centroids.
   - Assign each data point to the cluster with the nearest centroid.

3. **Update Step**:
   - Recalculate the centroid of each cluster by taking the mean of all data points assigned to that cluster.
   
4. **Repeat**: The assignment and update steps are repeated iteratively until convergence, i.e., until the centroids no longer change significantly, or a predefined number of iterations is reached.

5. **Convergence**: The algorithm converges when the assignments no longer change, or the change in centroids is below a certain threshold.

### Pseudocode

Here is a simplified pseudocode for the K-Means algorithm:

```plaintext
Initialize k centroids randomly
Repeat until convergence or for a set number of iterations:
    Assign each data point to the nearest centroid
    Update the centroids by calculating the mean of the data points assigned to each centroid
```

### Choosing the Number of Clusters \( k \)

Selecting the optimal number of clusters \( k \) is a crucial step in K-Means clustering. Common methods include:

- **Elbow Method**: Plot the sum of squared distances from each point to its assigned centroid as a function of \( k \). The optimal \( k \) is often found at the "elbow point" where the rate of decrease sharply slows down.
  
- **Silhouette Score**: This measures how similar a point is to its own cluster compared to other clusters. A higher silhouette score suggests better-defined clusters.

- **Gap Statistic**: Compares the total within intra-cluster variation for different values of \( k \) with their expected values under null reference distribution of the data.

### Advantages and Limitations

**Advantages**:
- Simple and easy to implement.
- Scales well to large datasets.
- Efficient in terms of computational cost, especially when using algorithms like K-Means++ for initialization.

**Limitations**:
- Requires the number of clusters \( k \) to be specified in advance.
- Sensitive to the initial placement of centroids, leading to potentially different clusters on different runs.
- Assumes clusters are spherical and evenly sized, which may not be the case in real data.
- Not suitable for identifying non-convex clusters.

### Applications

K-Means clustering is widely used in various fields, including:

- **Image Compression**: By clustering the colors in an image, K-Means can reduce the number of colors and thereby compress the image.
- **Customer Segmentation**: In marketing, customers can be clustered based on purchasing behavior, allowing targeted marketing strategies.
- **Document Clustering**: Grouping similar documents or text files based on content, which is useful in information retrieval and text mining.

### Variants and Extensions

- **K-Means++**: An improved version of the K-Means algorithm, where the initial centroids are chosen in a way that is likely to speed up convergence and improve clustering quality.
  
- **Mini-Batch K-Means**: A variant that uses small random samples (mini-batches) of the data to reduce computation time on very large datasets.

- **Fuzzy C-Means**: A generalization of K-Means that allows data points to belong to multiple clusters with varying degrees of membership.

- **Bisecting K-Means**: A hierarchical clustering method that repeatedly bisects clusters to create a hierarchical tree of clusters.

### Practical Considerations

1. **Data Preprocessing**: Before applying K-Means, itâ€™s often necessary to standardize the data, especially if the features have different scales.

2. **Dimensionality Reduction**: Techniques like PCA (Principal Component Analysis) are sometimes used to reduce the dimensionality of the data before applying K-Means, which can improve performance and interpretability.

# Demo Implementation

````python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

# For visualizing the elbow method
from yellowbrick.cluster import KElbowVisualizer
````

````python
# Generate Sample Data

# Generate synthetic dataset with 3 clusters
X, y = make_blobs(n_samples=300, centers=3, cluster_std=0.70, random_state=42)

# Standardize the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Plot the generated data
plt.scatter(X_scaled[:, 0], X_scaled[:, 1], s=50)
plt.title("Generated Data")
plt.show()
````

````python
# Applying K-Means Clustering

# Initialize KMeans with the number of clusters set to 3
kmeans = KMeans(n_clusters=3)

# Fit the model
kmeans.fit(X_scaled)

# Predict the clusters
y_kmeans = kmeans.predict(X_scaled)

# Plot the clustered data with centroids
plt.scatter(X_scaled[:, 0], X_scaled[:, 1], c=y_kmeans, s=50, cmap='viridis')

# Plot the centroids
centers = kmeans.cluster_centers_
plt.scatter(centers[:, 0], centers[:, 1], c='red', s=200, alpha=0.75, marker='x')
plt.title("K-Means Clustering")
plt.show()
````

````python
# Using the Elbow Method to Determine the Optimal Number of Clusters


# Initialize the KMeans model
model = KMeans()

# Use the Elbow method to find the optimal number of clusters
visualizer = KElbowVisualizer(model, k=(1,10))

visualizer.fit(X_scaled)        # Fit the data to the visualizer
visualizer.show()               # Finalize and render the figure
````

### Explanation of the Graph

The graph you've provided is an output from the **Elbow Method** used for determining the optimal number of clusters for K-Means clustering. The graph plots two key metrics against the number of clusters (\( k \)):

1. **Distortion Score (WCSS)**: This is plotted on the left Y-axis and represented by the blue line. It measures the within-cluster sum of squares (WCSS), which is the sum of squared distances between each point and its corresponding cluster centroid. Lower values of distortion indicate tighter clusters.

2. **Fit Time (Seconds)**: This is plotted on the right Y-axis and represented by the green dashed line. It shows the time taken to fit the K-Means model for each value of \( k \).

**Key Points on the Graph**:
- **X-Axis (k)**: Represents the number of clusters.
- **Y-Axis (left, distortion score)**: Represents the WCSS or distortion score, where lower values are better.
- **Y-Axis (right, fit time)**: Represents the time it took to fit the K-Means algorithm, where lower values are better for efficiency.

**The "Elbow" Point**:
- The graph identifies the "elbow" at \( k = 3 \) with a score of approximately 9.161.
- The elbow is where the distortion score sharply decreases before flattening out, which indicates that adding more clusters beyond this point does not significantly reduce the WCSS.

### Conclusion

**Optimal Number of Clusters**:
- The "elbow" is clearly observed at \( k = 3 \), which suggests that 3 is the optimal number of clusters for this dataset. This means that using 3 clusters will give you the best balance between the accuracy of the clustering (as indicated by the distortion score) and the simplicity of the model.

**Fit Time Consideration**:
- The fit time increases slightly with more clusters, but this isn't a major concern in this example. However, in larger datasets, the fit time could become more significant.

**Summary**:
- The Elbow Method graph suggests that you should choose 3 clusters for your K-Means model, as increasing beyond this point does not significantly improve the clustering quality. This is a good balance between model complexity and computational efficiency.

This conclusion aligns with typical clustering analysis practices, where the "elbow" point is used to determine the optimal number of clusters that adequately describe the data without overfitting.

- What is Within-Cluster Sum of Squares (WCSS)?
  - The Within-Cluster Sum of Squares (WCSS) is a measure of the total variance within each cluster. For each cluster, WCSS is the sum of the squared differences between each data point and the centroid of its cluster. The goal of K-Means is to minimize this value, as smaller WCSS indicates that the points are closer to the centroid and, therefore, more similar to each other within the cluster.

# Application 1: Patient Segmentation in Healthcare


In this application, we will implement K-Means clustering to segment patients based on certain health metrics. This kind of segmentation is crucial in personalized medicine, where treatment plans can be tailored according to the characteristics of each patient group.

Problem Statement:

We'll use a synthetic dataset that simulates patient data with attributes such as age, blood pressure, cholesterol level, and BMI (Body Mass Index). The goal is to segment the patients into distinct clusters, which might correspond to different risk profiles or treatment needs.

````python
# Import Necessary Libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from sklearn.datasets import make_blobs
import seaborn as sns

# For better visualization of clusters
sns.set(style="whitegrid")
````

````python
#  Generate or Load the Dataset

# Generate synthetic dataset with 4 clusters (simulating different patient profiles)
X, _ = make_blobs(n_samples=200, centers=4, cluster_std=1.5, random_state=42,
                  n_features=4)

# Creating a DataFrame for better understanding
columns = ['Age', 'Blood_Pressure', 'Cholesterol', 'BMI']
patients_df = pd.DataFrame(X, columns=columns)

# Displaying the first few rows of the dataset
patients_df.head()
````

````python
# Data Preprocessing

# Before applying K-Means, it is important to standardize the data because K-Means clustering uses the Euclidean distance, which is sensitive to the scale of the features.

# Standardizing the features
scaler = StandardScaler()
patients_scaled = scaler.fit_transform(patients_df)

# Converting the scaled data back into a DataFrame for convenience
patients_df_scaled = pd.DataFrame(patients_scaled, columns=columns)

# Displaying the first few rows of the scaled dataset
patients_df_scaled.head()
````

````python
#  Applying K-Means Clustering

# Applying K-Means with 4 clusters (since we simulated 4 profiles)
kmeans = KMeans(n_clusters=4, random_state=42)
# Fit the KMeans model to the data without the 'Cluster' column
patients_df_scaled['Cluster'] = kmeans.fit_predict(patients_df_scaled[columns])

# Display the first few rows with cluster assignments
patients_df_scaled.head()


#Note:
#We initialized the K-Means model with 4 clusters (based on our synthetic data).
#The fit_predict method fits the K-Means model to the data and returns the cluster labels for each data point, which are then appended to our DataFrame.
````

````python
# Visualizing the Clusters

#Since we have a 4-dimensional dataset, visualizing it directly is challenging. We can use pair plots to visualize the relationships between features and how clusters are formed:


# Pairplot to visualize clusters
sns.pairplot(patients_df_scaled, hue='Cluster', palette='viridis', diag_kind='kde')
plt.suptitle("Patient Clusters", y=1.02)
plt.show()

# This pair plot shows the pairwise relationships between features and colors the points according to their cluster assignments.
````

````python
# Calculate the centroid of each cluster in the original scale
centroids = scaler.inverse_transform(kmeans.cluster_centers_)

# Create a DataFrame for centroids
centroids_df = pd.DataFrame(centroids, columns=columns)

# Display the centroids
centroids_df
````

# Application 2: Recommendation System Using K-Means Clustering

We will use a publicly available dataset from MovieLens, which contains ratings data for movies.

The goal is to cluster users based on their movie ratings and then recommend movies to users based on the cluster they belong to.

````python
# Importing Necessary Libraries

import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
from sklearn.metrics import pairwise_distances_argmin_min
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
````

````python
# Loading and Exploring the Dataset

# Use 'MovieLens' 100k dataset, which contains 100,000 ratings from 943 users on 1682 movies.

# Load the dataset
url = 'http://files.grouplens.org/datasets/movielens/ml-100k/u.data'
columns = ['user_id', 'item_id', 'rating', 'timestamp']
df = pd.read_csv(url, sep='\t', names=columns)

# Display the first few rows of the dataset
df.head()
````

````python
# Data Preprocessing

# We'll transform the dataset into a user-item matrix where rows represent users, columns represent movies, and the values are the ratings.

# Create a user-item matrix
user_item_matrix = df.pivot_table(index='user_id', columns='item_id', values='rating')

# Fill NaN values with 0 (indicating that a user has not rated a movie)
user_item_matrix.fillna(0, inplace=True)

# Standardize the features
scaler = StandardScaler()
user_item_matrix_scaled = scaler.fit_transform(user_item_matrix)

# Check the shape of the data
user_item_matrix_scaled.shape
````

````python
#  Dimensionality Reduction

# To improve the performance of the K-Means algorithm and make the visualization easier, we'll reduce the dimensionality of the data using PCA.

# Apply PCA to reduce dimensionality
pca = PCA(n_components=2)
user_item_matrix_pca = pca.fit_transform(user_item_matrix_scaled)

# Check the explained variance
print(f"Explained variance by components: {pca.explained_variance_ratio_}")
````

````python
# Applying K-Means Clustering
# Apply K-Means clustering
kmeans = KMeans(n_clusters=10, random_state=42)
kmeans.fit(user_item_matrix_pca)

# Assign the cluster labels to each user
user_item_matrix['cluster'] = kmeans.labels_
````

````python
# Visualizing the Clusters

# Plot the clusters
plt.figure(figsize=(10, 8))
plt.scatter(user_item_matrix_pca[:, 0], user_item_matrix_pca[:, 1], c=kmeans.labels_, s=50, cmap='viridis')
plt.title('User Clusters based on Movie Ratings')
plt.xlabel('PCA Component 1')
plt.ylabel('PCA Component 2')
plt.show()
````

````python
# Making Recommendations

# To recommend movies to a user, we'll select a user from a cluster and recommend the most popular movies within that cluster that the user hasn't rated yet.

# Function to recommend movies to a user based on their cluster
def recommend_movies(user_id, user_item_matrix, n_recommendations=5):
    # Get the user's cluster
    user_cluster = user_item_matrix.loc[user_id, 'cluster']

    # Get all users in the same cluster
    cluster_users = user_item_matrix[user_item_matrix['cluster'] == user_cluster]

    # Calculate the average ratings of each movie in this cluster
    cluster_movie_ratings = cluster_users.drop(columns=['cluster']).mean().sort_values(ascending=False)

    # Get the movies that the user has not rated yet
    user_ratings = user_item_matrix.loc[user_id].drop('cluster')
    unrated_movies = user_ratings[user_ratings == 0]

    # Recommend the top N unrated movies
    recommendations = cluster_movie_ratings[unrated_movies.index].head(n_recommendations)

    return recommendations

# Example: Recommend movies to user with ID 100
recommendations = recommend_movies(100, user_item_matrix)
print(recommendations)
````

# Application 3: Supply Chain Optimization

To implement supply chain optimization using K-Means clustering, we can focus on clustering suppliers or logistics routes based on various performance metrics like delivery time, cost, reliability, etc.


````python
# importing libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import silhouette_score
import seaborn as sns

# Set the random seed for reproducibility
np.random.seed(42)
````

````python
#Generate a Synthetic Dataset

#For demonstration purposes, we will create a synthetic dataset that simulates supplier performance metrics. Let's assume the dataset includes the following features for each supplier:
  # - Delivery Time: The average time (in days) it takes for the supplier to deliver goods.
  # - Cost: The average cost per unit of goods supplied.
  # - Reliability: A score representing the reliability of the supplier (e.g., percentage of on-time deliveries).
  # - Order Volume: The average order volume handled by the supplier.


# Create a synthetic dataset
n_suppliers = 1000

# Generate data
data = {
    'Supplier_ID': range(1, n_suppliers + 1),
    'Delivery_Time': np.random.normal(10, 2, n_suppliers),  # days
    'Cost': np.random.normal(500, 100, n_suppliers),        # cost per unit
    'Reliability': np.random.uniform(70, 100, n_suppliers), # reliability percentage
    'Order_Volume': np.random.normal(1000, 200, n_suppliers) # units
}

# Convert to DataFrame
df = pd.DataFrame(data)
df.head()

````

````python
#  Data Preprocessing


# Standardize the features
scaler = StandardScaler()
scaled_features = scaler.fit_transform(df[['Delivery_Time', 'Cost', 'Reliability', 'Order_Volume']])

# Convert back to a DataFrame for easier interpretation
scaled_df = pd.DataFrame(scaled_features, columns=['Delivery_Time', 'Cost', 'Reliability', 'Order_Volume'])
scaled_df.head()
````

````python
# Apply K-Means Clustering

# Determine the optimal number of clusters using the Elbow Method
wcss = []  # Within-cluster sum of squares

# Test for a range of cluster numbers
for k in range(1, 11):
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(scaled_df)
    wcss.append(kmeans.inertia_)  # Inertia: WCSS for the current k

# Plot the Elbow Method graph
plt.plot(range(1, 11), wcss, marker='o', linestyle='--')
plt.title('Elbow Method for Optimal k')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
````

````python
# Evaluate and Choose the Number of Clusters

# Apply K-Means with the chosen number of clusters
kmeans = KMeans(n_clusters=4, random_state=42) # 4 is taken for demonstration
kmeans.fit(scaled_df)

# Add the cluster labels to the original DataFrame
df['Cluster'] = kmeans.labels_
````

````python
# Analyze the Clusters

# Analyze the cluster centers
centers = pd.DataFrame(scaler.inverse_transform(kmeans.cluster_centers_),
                       columns=['Delivery_Time', 'Cost', 'Reliability', 'Order_Volume'])
centers['Cluster'] = range(1, 5)
centers = centers.set_index('Cluster')
centers

# Plot the cluster characteristics
sns.pairplot(df, hue='Cluster', palette='viridis')
plt.show()

# Count the number of suppliers in each cluster
df['Cluster'].value_counts()
````
