# Isolation Forest

#### Overview
Isolation Forest (iForest) is an unsupervised learning algorithm primarily used for anomaly detection. It is based on the idea that anomalies are "few and different." Unlike other methods that rely on distance or density, the Isolation Forest algorithm isolates observations by randomly selecting a feature and then randomly selecting a split value between the maximum and minimum values of the selected feature. This process creates an isolation tree (iTree), and the path length from the root of the tree to a particular data point can be used to determine whether it is an anomaly.

#### Key Concepts

1. **Isolation**:
   - The core idea of the Isolation Forest is that outliers or anomalies can be isolated quickly, using fewer random partitions, because they are few and different.
   - The algorithm isolates observations by recursively splitting data points in a way that shorter paths indicate anomalies.
   
2. **Isolation Tree (iTree)**:
   - An isolation tree is a binary tree structure where each node represents a feature and a random split point.
   - For each node, the algorithm randomly selects a feature and a split value within the range of that feature.
   - The process continues until the node contains only one observation or until all observations are isolated.
   - The depth of the tree corresponds to the number of splits required to isolate a point.

4. **Forest of Trees**:
   - The Isolation Forest consists of multiple isolation trees. Each tree is constructed using a random subset of the data and features, which helps ensure that the algorithm does not become biased by any particular feature.
   - The final anomaly score is an aggregate measure, usually the average score across all trees.

#### Process of Isolation Forest

1. **Subsampling**:
   - A random subset of the data is selected to build an individual iTree.
   - Subsampling improves the efficiency and robustness of the algorithm.

2. **Random Feature Selection**:
   - For each node, a feature is randomly selected from the available features.
   - This random selection helps in isolating observations faster.

3. **Random Split Selection**:
   - A split value is chosen randomly between the maximum and minimum values of the selected feature.
   - This step ensures that anomalies, which often lie at the fringes of the data distribution, are isolated quickly.

4. **Tree Construction**:
   - The process of random feature and split selection continues until each observation is isolated, resulting in the construction of an iTree.
   - The depth of the tree indicates how easily an observation can be isolated.

5. **Scoring**:
   - Once all trees are constructed, each observation is passed through the trees to compute its path length.
   - The average path length across all trees is used to calculate the anomaly score.

#### Advantages of Isolation Forest

- **Efficiency**: The algorithm is computationally efficient, making it suitable for large datasets.
- **Scalability**: It scales well with both the number of observations and features.
- **No Assumptions**: Unlike distance-based methods, Isolation Forest does not require the calculation of pairwise distances, nor does it assume any underlying data distribution.
- **Versatility**: It works well in both low and high-dimensional spaces.

#### Disadvantages

- **Randomness**: The algorithm's performance can be influenced by the randomness in feature and split selection, although this can be mitigated by averaging across many trees.
- **Interpretability**: The results, particularly the anomaly scores, can be less interpretable compared to density-based methods or clustering.

# Application 3: Healthcare and Medical Diagnostics


we'll simulate a scenario where we detect anomalies in medical data, such as patient vital signs. We'll assume that the dataset consists of vital signs like heart rate, blood pressure, and body temperature. We'll use Isolation Forest to identify any abnormal readings that could indicate potential health issues.

```python
import numpy as np
import pandas as pd
from sklearn.ensemble import IsolationForest
import matplotlib.pyplot as plt

# For reproducibility
np.random.seed(42)
```

```python
# Generate Simulated Medical Data

# Simulate normal medical data
n_samples = 1000
normal_data = {
    'heart_rate': np.random.normal(70, 10, n_samples),  # Normal heart rate around 70 bpm
    'blood_pressure': np.random.normal(120, 15, n_samples),  # Normal BP around 120 mmHg
    'body_temperature': np.random.normal(37, 0.5, n_samples)  # Normal temperature around 37°C
}

# Convert to DataFrame
df = pd.DataFrame(normal_data)

# Inject some anomalies
n_anomalies = 50
anomalies = {
    'heart_rate': np.random.normal(120, 5, n_anomalies),  # Anomalous heart rate
    'blood_pressure': np.random.normal(180, 10, n_anomalies),  # Anomalous BP
    'body_temperature': np.random.normal(39, 0.2, n_anomalies)  # Anomalous temperature
}

# Add anomalies to the dataset
df_anomalies = pd.DataFrame(anomalies)
df = pd.concat([df, df_anomalies], ignore_index=True) # Use concat instead of append

# Shuffle the data
df = df.sample(frac=1).reset_index(drop=True)

# Display the first few rows
df.head()
```

```python
#  Apply Isolation Forest for Anomaly Detection

# Initialize the Isolation Forest model
model = IsolationForest(n_estimators=100, contamination=float(n_anomalies) / n_samples, random_state=42)

# Fit the model
model.fit(df)

# Predict anomalies (1 = normal, -1 = anomaly)
df['anomaly'] = model.predict(df)

# Convert prediction (-1 for anomaly) to a boolean for easier analysis
df['anomaly'] = df['anomaly'].apply(lambda x: 1 if x == -1 else 0)

# Count anomalies detected
print("Number of anomalies detected: ", df['anomaly'].sum())
```

```python
# Analyze and Visualize the Results

# Visualize anomalies
fig, axs = plt.subplots(3, 1, figsize=(10, 15))

axs[0].scatter(df.index, df['heart_rate'], c=df['anomaly'], cmap='coolwarm')
axs[0].set_title('Heart Rate Anomalies')
axs[0].set_ylabel('Heart Rate (bpm)')

axs[1].scatter(df.index, df['blood_pressure'], c=df['anomaly'], cmap='coolwarm')
axs[1].set_title('Blood Pressure Anomalies')
axs[1].set_ylabel('Blood Pressure (mmHg)')

axs[2].scatter(df.index, df['body_temperature'], c=df['anomaly'], cmap='coolwarm')
axs[2].set_title('Body Temperature Anomalies')
axs[2].set_ylabel('Body Temperature (°C)')

plt.show()
```

```python
# Interpretation of Results
# Display a few examples of anomalies detected
anomalies_detected = df[df['anomaly'] == 1]
print("Examples of detected anomalies:")
print(anomalies_detected.head())

# Summary statistics
normal_data = df[df['anomaly'] == 0]
anomalous_data = df[df['anomaly'] == 1]

print("\nNormal Data Statistics:")
print(normal_data.describe())

print("\nAnomalous Data Statistics:")
print(anomalous_data.describe())

```

- **Observation**: The majority of the data points (blue) cluster around the normal heart rate range (approximately 60 to 100 beats per minute). The anomalies (red dots) are primarily in the higher range, around 120 bpm.
- **Conclusion**: The Isolation Forest successfully identified heart rate readings that are significantly higher than normal, indicating potential health issues such as tachycardia or other cardiovascular conditions.

#### **Blood Pressure Anomalies:**
- **Visualization**: The second plot represents blood pressure readings. Again, blue dots represent normal readings, while red dots indicate anomalies.
- **Observation**: Most of the normal data points are clustered around the typical range of 100 to 140 mmHg systolic pressure. The anomalies are mainly in the higher range, close to 180-200 mmHg, but there are also a few lower than 100 mmHg.
- **Conclusion**: The model effectively detected blood pressure anomalies that may indicate hypertension or hypotension, both of which require medical attention. The anomalies on the lower side might also suggest hypotension, which can be dangerous if not monitored.

#### **Body Temperature Anomalies:**
- **Visualization**: The third plot shows body temperature readings. Normal readings are in blue, while anomalies are in red.
- **Observation**: The normal body temperature data points are clustered around 36.5°C to 37.5°C. The anomalies detected by the model are primarily in the higher range (around 39°C), indicating a potential fever.
- **Conclusion**: The model accurately detected high body temperature readings, which could indicate fever or an infection. These anomalies could be critical for early detection of illness.
