# Hierarchical Clustering

Hierarchical clustering is a method of cluster analysis that seeks to build a hierarchy of clusters. It is an unsupervised learning algorithm, which means it is used to group data into clusters without predefined labels or classes. This technique is particularly useful when you want to discover the inherent structure in data.

#### Key Concepts

1. **Hierarchy of Clusters**:
   - Hierarchical clustering creates a tree-like structure called a **dendrogram**, which represents the nested grouping of data points and the levels at which these groupings occur.
   - The root of the dendrogram represents a single cluster containing all data points, and the leaves represent individual data points. Intermediate levels represent clusters formed by merging or splitting clusters.

2. **Types of Hierarchical Clustering**:
   - **Agglomerative (Bottom-Up)**: This is the most common type. It starts with each data point as a separate cluster and merges the closest pairs of clusters iteratively until all points are in one cluster.
   - **Divisive (Top-Down)**: The opposite of agglomerative, this method starts with all data points in a single cluster and recursively splits them into smaller clusters.

3. **Distance Measures**:
   - The distance or dissimilarity between data points is a crucial part of hierarchical clustering. Common distance measures include:
     - **Euclidean Distance**: The straight-line distance between two points in a multi-dimensional space.
     - **Manhattan Distance**: The sum of the absolute differences between the coordinates of the points.
     - **Cosine Similarity**: Measures the cosine of the angle between two vectors, often used in text clustering.
     - **Minkowski Distance**: A generalization of both Euclidean and Manhattan distances.

4. **Linkage Criteria**:
   - Determines how the distance between clusters is calculated. Common methods include:
     - **Single Linkage (Nearest Neighbor)**: The distance between the closest points of two clusters.
     - **Complete Linkage (Farthest Neighbor)**: The distance between the farthest points in two clusters.
     - **Average Linkage**: The average distance between all pairs of points in two clusters.
     - **Centroid Linkage**: The distance between the centroids (mean points) of two clusters.
     - **Ward’s Method**: Minimizes the variance within clusters, leading to compact and spherical clusters.

5. **Dendrogram**:
   - A dendrogram is a visual representation of the hierarchical clustering process. It shows at what level of similarity or distance clusters are combined.
   - By cutting the dendrogram at a specific level, you can choose a number of clusters that make sense for your data.

#### Steps in Hierarchical Clustering

1. **Calculate the Distance Matrix**:
   - Start by calculating the distance between each pair of data points using a chosen distance measure, creating a distance matrix.

2. **Start Clustering**:
   - For agglomerative clustering, start by treating each data point as a single cluster.

3. **Merge or Split Clusters**:
   - Use the chosen linkage criterion to find the closest pair of clusters and merge them into a single cluster. This step is repeated until all points are merged into a single cluster (for agglomerative) or all points are separated into individual clusters (for divisive).

4. **Create Dendrogram**:
   - As clusters are merged or split, the dendrogram is updated to reflect these changes, showing the hierarchy of clusters.

5. **Determine the Number of Clusters**:
   - Once the dendrogram is complete, you can decide on the optimal number of clusters by cutting the dendrogram at the desired level.

#### Advantages of Hierarchical Clustering

- **No need to specify the number of clusters upfront**: Unlike K-means clustering, you don’t need to specify the number of clusters at the beginning.
- **Hierarchical structure**: Provides a clear hierarchy of clusters, which can be more informative than just a flat clustering.
- **Versatile**: Can be used for both small datasets and large datasets with some optimization.

#### Disadvantages of Hierarchical Clustering

- **Computationally expensive**: Hierarchical clustering can be slow and memory-intensive, especially with large datasets, because it requires the computation and storage of an entire distance matrix.
- **Sensitive to noise and outliers**: Outliers can disproportionately affect the clustering process.
- **Once merged or split, clusters cannot be undone**: This can lead to a less optimal clustering result in some cases.

#### Applications of Hierarchical Clustering

- **Gene expression data analysis**: Grouping similar gene expression profiles.
- **Market segmentation**: Identifying distinct customer groups based on purchasing behavior.
- **Document classification**: Organizing a large set of documents into a hierarchy of topics.
- **Social network analysis**: Discovering communities within social networks.

#### Hierarchical Clustering in Practice

- **Scalability**: While hierarchical clustering can be computationally intensive, optimizations such as using KD-trees or Ball-trees for distance calculations, and using clustering methods like BIRCH, can help scale hierarchical clustering to larger datasets.
- **Software and Tools**: Hierarchical clustering is implemented in many data analysis tools and libraries, including Python’s `scipy` and `scikit-learn`, R’s `hclust` function, and others.

# Application 1:  Image Segmentation

```python
# Import libraries

import numpy as np
from sklearn.cluster import AgglomerativeClustering
from sklearn.utils import shuffle
from PIL import Image
import matplotlib.pyplot as plt
```

```python
#  Load and preprocess the image
def load_image(image_path):
    image = Image.open(image_path)
    image = image.resize((100, 100))  # Resize for faster processing
    image_np = np.array(image)
    return image_np
```

```python
# Reshape the image data for clustering
def preprocess_image(image_np):
    w, h, d = image_np.shape
    image_array = np.reshape(image_np, (w * h, d))
    return image_array
```

```python
#  Apply hierarchical clustering
def hierarchical_clustering(image_array, n_clusters):
    # Removed sampling for accurate clustering
    model = AgglomerativeClustering(n_clusters=n_clusters)
    labels = model.fit_predict(image_array)
    return labels, model
```

```python
# Replace each pixel color with the mean color of its cluster
def create_segmented_image(image_array, labels, model, w, h):
    # Compute the mean color for each cluster
    mean_colors = np.zeros((model.n_clusters_, image_array.shape[1]))
    for i in range(model.n_clusters_):
        mean_colors[i] = np.mean(image_array[labels == i], axis=0)


    segmented_image = np.copy(image_array)
    for i in range(model.n_clusters_):
        segmented_image[labels == i] = mean_colors[i]

    segmented_image = np.reshape(segmented_image, (w, h, image_array.shape[1]))
    return segmented_image
```

```python
# Visualize the segmented image
def display_images(original, segmented):
    fig, ax = plt.subplots(1, 2, figsize=(12, 6))
    ax[0].imshow(original)
    ax[0].set_title("Original Image")
    ax[0].axis('off')

    ax[1].imshow(segmented.astype(np.uint8))
    ax[1].set_title("Segmented Image")
    ax[1].axis('off')

    plt.show()
```

```python
# Load and process the image
image_path = 'image.jpg'  # Replace with your image path
image_np = load_image(image_path)
image_array = preprocess_image(image_np)
w, h, d = image_np.shape

# Perform hierarchical clustering
n_clusters = 5  # Number of segments
labels, model = hierarchical_clustering(image_array, n_clusters)

# Create segmented image
segmented_image = create_segmented_image(image_array, labels, model, w, h)

# Display the original and segmented images
display_images(image_np, segmented_image)
```

# Application 2: Student Performance Analysis

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from scipy.cluster.hierarchy import dendrogram, linkage, fcluster
```

```python
# Sample data: Students' scores in Math, English, and Science
data = {
    'Student': ['Alice', 'Bob', 'Charlie', 'David', 'Eva', 'Frank', 'Grace', 'Hannah', 'Ivy', 'Jack'],
    'Math': [88, 92, 85, 70, 65, 77, 90, 68, 74, 81],
    'English': [78, 85, 80, 72, 60, 66, 88, 75, 70, 73],
    'Science': [90, 89, 82, 88, 60, 79, 95, 64, 77, 80]
}

# Convert to DataFrame
df = pd.DataFrame(data)
print(df)
```

```python
# Standardize the data
scaler = StandardScaler()
X = df[['Math', 'English', 'Science']]
X_scaled = scaler.fit_transform(X)
```

```python
# Perform hierarchical clustering
Z = linkage(X_scaled, method='ward')

# Plot the dendrogram
plt.figure(figsize=(10, 7))
dendrogram(Z, labels=df['Student'].values, leaf_rotation=90, leaf_font_size=10)
plt.title("Dendrogram for Student Performance")
plt.xlabel("Students")
plt.ylabel("Euclidean Distance")
plt.show()
# The dendrogram will visually show the hierarchical relationships between students based on their performance in the three subjects.
```

```python
# Set the threshold for the dendrogram cut
max_d = 6  # This can be adjusted based on the dendrogram

# Generate cluster labels
clusters = fcluster(Z, max_d, criterion='distance')

# Add the cluster labels to the original dataframe
df['Cluster'] = clusters

# Display the dataframe with clusters
print(df)
```

```python
# Display students grouped by cluster
for cluster in sorted(df['Cluster'].unique()):
    print(f"\nCluster {cluster}:")
    print(df[df['Cluster'] == cluster][['Student', 'Math', 'English', 'Science']])
```

```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
from scipy.cluster.hierarchy import dendrogram, linkage, fcluster
import matplotlib.pyplot as plt
```

```python
# Create a synthetic dataset representing financial transactions:

np.random.seed(42)

# Generate synthetic data
num_transactions = 100
data = {
    'TransactionAmount': np.random.normal(100, 50, num_transactions),
    'TransactionTime': np.random.uniform(0, 24, num_transactions),
    'IsFraudulent': np.random.choice([0, 1], size=num_transactions, p=[0.95, 0.05])
}

df = pd.DataFrame(data)

# Introduce some anomalies to simulate fraudulent transactions
df.loc[df['IsFraudulent'] == 1, 'TransactionAmount'] = df['TransactionAmount'] * np.random.uniform(2, 10)
df.loc[df['IsFraudulent'] == 1, 'TransactionTime'] = df['TransactionTime'] + np.random.uniform(-5, 5)

print(df.head())

```

```python
# preprocessing the data

# Drop the 'IsFraudulent' column for clustering
X = df.drop('IsFraudulent', axis=1)

# Standardize the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
# Perform hierarchical clustering
linked = linkage(X_scaled, method='ward')

# Plot the dendrogram
plt.figure(figsize=(10, 7))
dendrogram(linked,
           orientation='top',
           distance_sort='descending',
           show_leaf_counts=True)
plt.title('Dendrogram for Financial Transactions')
plt.xlabel('Transaction Index')
plt.ylabel('Euclidean Distance')
plt.show()
```

```python
# Create clusters
clusters = fcluster(linked, t=5, criterion='distance')

# Add cluster labels to the original dataframe
df['Cluster'] = clusters

print(df.head())
```
